```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function and its gradient
def cost_function(theta):
    return theta ** 2

def gradient(theta):
    return 2 * theta

# Batch Gradient Descent
def batch_gradient_descent(theta_init, learning_rate, num_iterations):
    theta = theta_init
    history = [theta]
    for _ in range(num_iterations):
        grad = gradient(theta)
        theta = theta - learning_rate * grad
        history.append(theta)
    return history

# Stochastic Gradient Descent (with more pronounced randomness)
def stochastic_gradient_descent(theta_init, learning_rate, num_iterations):
    theta = theta_init
    history = [theta]
    for _ in range(num_iterations):
        grad = gradient(theta)
        noise = np.random.uniform(0.5, 1.5)
        theta = theta - learning_rate * grad * noise
        history.append(theta)
    return history

# Mini-batch Gradient Descent (add batch variability)
def mini_batch_gradient_descent(theta_init, learning_rate, num_iterations, batch_size):
    theta = theta_init
    history = [theta]
    for _ in range(num_iterations):
        grad = gradient(theta)
        batch_factor = np.random.uniform(0.8, 1.2, batch_size).mean()
        theta = theta - learning_rate * grad * batch_factor
        history.append(theta)
    return history

# Plot function for subplots
def plot_subplots(batch_history, sgd_history, mini_batch_history):
    theta_range = np.linspace(-5, 5, 100)
    
    # Create subplots
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))
    
    # Batch Gradient Descent
    axs[0].plot(theta_range, cost_function(theta_range), label="Cost Function (y = x^2)", color="black", linestyle="--")
    axs[0].plot(batch_history, cost_function(np.array(batch_history)), marker="o", color="red", label="Batch Gradient Descent")
    axs[0].set_title("Batch Gradient Descent")
    axs[0].set_xlabel(r"$\theta$")
    axs[0].set_ylabel(r"Cost ($\theta^2$)")
    axs[0].legend()
    axs[0].grid(True)
    
    # Stochastic Gradient Descent
    axs[1].plot(theta_range, cost_function(theta_range), label="Cost Function (y = x^2)", color="black", linestyle="--")
    axs[1].plot(sgd_history, cost_function(np.array(sgd_history)), marker="o", color="green", label="Stochastic Gradient Descent")
    axs[1].set_title("Stochastic Gradient Descent")
    axs[1].set_xlabel(r"$\theta$")
    axs[1].legend()
    axs[1].grid(True)
    
    # Mini-batch Gradient Descent
    axs[2].plot(theta_range, cost_function(theta_range), label="Cost Function (y = x^2)", color="black", linestyle="--")
    axs[2].plot(mini_batch_history, cost_function(np.array(mini_batch_history)), marker="o", color="yellow", label="Mini-batch Gradient Descent")
    axs[2].set_title("Mini-batch Gradient Descent")
    axs[2].set_xlabel(r"$\theta$")
    axs[2].legend()
    axs[2].grid(True)
    
    # Display the subplots
    plt.tight_layout()
    plt.show()

# Initialize parameters
theta_init = 4
learning_rate = 0.1
num_iterations = 15
batch_size = 3

# Run each gradient descent technique
batch_history = batch_gradient_descent(theta_init, learning_rate, num_iterations)
sgd_history = stochastic_gradient_descent(theta_init, learning_rate, num_iterations)
mini_batch_history = mini_batch_gradient_descent(theta_init, learning_rate, num_iterations, batch_size)

# Plot the subplots with common color
plot_subplots(batch_history, sgd_history, mini_batch_history)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function and its gradient
def cost_function(theta):
    return theta ** 2

def gradient(theta):
    return 2 * theta

# Batch Gradient Descent
def batch_gradient_descent(theta_init, learning_rate, num_iterations):
    theta = theta_init
    history = [theta]
    for _ in range(num_iterations):
        grad = gradient(theta)
        theta = theta - learning_rate * grad
        history.append(theta)
    return history

# Stochastic Gradient Descent (with randomness)
def stochastic_gradient_descent(theta_init, learning_rate, num_iterations):
    theta = theta_init
    history = [theta]
    for _ in range(num_iterations):
        grad = gradient(theta)
        noise = np.random.uniform(0.5, 1.5)
        theta = theta - learning_rate * grad * noise
        history.append(theta)
    return history

# Mini-batch Gradient Descent (with batch randomness)
def mini_batch_gradient_descent(theta_init, learning_rate, num_iterations, batch_size):
    theta = theta_init
    history = [theta]
    for _ in range(num_iterations):
        grad = gradient(theta)
        batch_factor = np.random.uniform(0.8, 1.2, batch_size).mean()
        theta = theta - learning_rate * grad * batch_factor
        history.append(theta)
    return history

# Plot function for superimposed graph
def plot_superimposed_graph(batch_history, sgd_history, mini_batch_history):
    theta_range = np.linspace(-5, 5, 100)
    
    # Plot the cost function
    plt.plot(theta_range, cost_function(theta_range), label="Cost Function (y = x^2)", color="black", linestyle="--")
    
    # Plot gradient descent paths
    plt.plot(batch_history, cost_function(np.array(batch_history)), marker="o", color="red", label="Batch Gradient Descent")
    plt.plot(sgd_history, cost_function(np.array(sgd_history)), marker="o", color="green", label="Stochastic Gradient Descent")
    plt.plot(mini_batch_history, cost_function(np.array(mini_batch_history)), marker="o", color="yellow", label="Mini-batch Gradient Descent")
    
    # Title and labels
    plt.title("Comparison of Gradient Descent Techniques")
    plt.xlabel(r"$\theta$")
    plt.ylabel(r"Cost ($\theta^2$)")
    
    # Legend and grid
    plt.legend()
    plt.grid(True)
    
    # Show plot
    plt.show()

# Initialize parameters
theta_init = 4
learning_rate = 0.1
num_iterations = 15
batch_size = 3

# Run each gradient descent technique
batch_history = batch_gradient_descent(theta_init, learning_rate, num_iterations)
sgd_history = stochastic_gradient_descent(theta_init, learning_rate, num_iterations)
mini_batch_history = mini_batch_gradient_descent(theta_init, learning_rate, num_iterations, batch_size)

# Plot the superimposed graph
plot_superimposed_graph(batch_history, sgd_history, mini_batch_history)
```
